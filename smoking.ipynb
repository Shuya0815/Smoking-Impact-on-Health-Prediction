```python
from skrub import TableVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader
import torch
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score, recall_score, precision_score

# Device configuration
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Data Preprocessing
# Identify numeric columns for scaling
cols_to_scale = smoking_df.drop(['smoking', 'ID'], axis=1).select_dtypes(include='number').columns.tolist()

# Use TableVectorizer with StandardScaler on numeric columns
tv = TableVectorizer(specific_transformers=[(StandardScaler(), cols_to_scale)])
X = tv.fit_transform(smoking_df.drop(['smoking', 'ID'], axis=1))
y = smoking_df['smoking']

if isinstance(X, pd.DataFrame):
    X = X.to_numpy()

# Train/Validation/Test Split
xtrain, xtemp, ytrain, ytemp = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
xval, xtest, yval, ytest = train_test_split(xtemp, ytemp, test_size=0.5, stratify=ytemp, random_state=42)

# Convert to numpy arrays if needed
if isinstance(xtrain, pd.DataFrame):
    xtrain = xtrain.to_numpy()
    xval = xval.to_numpy()
    xtest = xtest.to_numpy()

# Convert to PyTorch tensors
xtrain = torch.tensor(xtrain, dtype=torch.float32, device=DEVICE)
ytrain = torch.tensor(ytrain.values, dtype=torch.float32, device=DEVICE)
xval = torch.tensor(xval, dtype=torch.float32, device=DEVICE)
yval = torch.tensor(yval.values, dtype=torch.float32, device=DEVICE)
xtest = torch.tensor(xtest, dtype=torch.float32, device=DEVICE)
ytest = torch.tensor(ytest.values, dtype=torch.float32, device=DEVICE)

# Create TensorDataset and DataLoader
train_data = TensorDataset(xtrain, ytrain)
val_data = TensorDataset(xval, yval)

train_loader = DataLoader(train_data, batch_size=128, shuffle=True, drop_last=True)
val_loader = DataLoader(val_data, batch_size=64, shuffle=False, drop_last=False)

# Define the Model
class TorchClassifier(torch.nn.Module):
    def __init__(self, n_input, n_output):
        super().__init__()
        self.fcl = torch.nn.Sequential(
            torch.nn.Linear(n_input, 512), torch.nn.ReLU(), torch.nn.Dropout(0.2),
            torch.nn.Linear(512, 256), torch.nn.ReLU(), torch.nn.Dropout(0.2),
            torch.nn.Linear(256, 128), torch.nn.ReLU(), torch.nn.Dropout(0.2),
            torch.nn.Linear(128, 64), torch.nn.ReLU(), torch.nn.Dropout(0.2),
            torch.nn.Linear(64, 32), torch.nn.ReLU(), torch.nn.Dropout(0.2),
            torch.nn.Linear(32, n_output)
        )

    def forward(self, x):
        return self.fcl(x)

# Initialize the Model
model = TorchClassifier(xtrain.shape[1], 1).to(DEVICE)
opt = torch.optim.Adam(model.parameters(), lr=0.001)
n_epochs = 500
model_perf_hist = pd.DataFrame(columns=['EPOCH', 'STATE', 'LOSS', 'ACCURACY', 'PRECISION', 'RECALL', 'F1'])

# Training Loop
model.train()
for epoch in range(n_epochs):
    train_correct, train_samples = 0, 0
    all_preds = []
    all_targets = []
    final_train_loss = None

    for xbatch, ybatch in train_loader:
        # Forward pass
        train_logits = model(xbatch).squeeze()
        train_loss = torch.nn.functional.binary_cross_entropy_with_logits(train_logits, ybatch)
        
        # Compute predictions
        train_preds = torch.sigmoid(train_logits)
        train_binary_preds = (train_preds > 0.5).float()

        # Update accuracy
        train_correct += (train_binary_preds == ybatch).sum().item()
        train_samples += ybatch.size(0)

        # Collect predictions and targets
        all_preds.append(train_binary_preds.detach().cpu().numpy())
        all_targets.append(ybatch.detach().cpu().numpy())

        # Backward pass
        opt.zero_grad()
        train_loss.backward()
        opt.step()

        final_train_loss = train_loss.item()

    # Compute full-epoch metrics
    all_preds = np.concatenate(all_preds)
    all_targets = np.concatenate(all_targets)

    train_accuracy = round(train_correct / train_samples * 100, 2)
    train_precision = precision_score(all_targets, all_preds)
    train_recall = recall_score(all_targets, all_preds)
    train_f1 = f1_score(all_targets, all_preds)

    # Store metrics
    model_perf_hist.loc[epoch] = {
        'EPOCH': epoch,
        'STATE': 'train',
        'LOSS': final_train_loss,
        'ACCURACY': train_accuracy,
        'PRECISION': train_precision,
        'RECALL': train_recall,
        'F1': train_f1
    }

    if epoch % 10 == 0:
        print(
            f'EPOCH: {epoch} | TRAIN LOSS: {final_train_loss:.4f} | '
            f'ACCURACY: {train_accuracy:.2f}% | PRECISION: {train_precision:.4f} | '
            f'RECALL: {train_recall:.4f} | F1: {train_f1:.4f}'
        )

# Save Training History
model_perf_hist.to_csv('model_performance_history.csv', index=False)
```
